## 2.1. ANÁLISIS DE VISITAS

Se analizan las métricas relacionadas con el uso del sitio web: número de usuarios y sesiones a lo largo del tiempo, duración de las sesiones y frecuencia de retorno. Este análisis permite entender cómo los usuarios interactúan con el servicio y con qué regularidad regresan.

### 2.1.1. Usuarios activos por período (DAU, WAU, MAU)

```python
# Crear columnas temporales para agrupación
visits['date'] = visits['start_ts'].dt.date
visits['week'] = visits['start_ts'].dt.to_period('W').dt.to_timestamp()
visits['month'] = visits['start_ts'].astype('datetime64[M]')

# DAU: Usuarios activos diarios
dau = (
    visits
    .groupby('date')['uid']
    .nunique()
    .reset_index(name='daily_users')
)

# WAU: Usuarios activos semanales
wau = (
    visits
    .groupby('week')['uid']
    .nunique()
    .reset_index(name='weekly_users')
)

# MAU: Usuarios activos mensuales
mau = (
    visits
    .groupby('month')['uid']
    .nunique()
    .reset_index(name='monthly_users')
)

# Resumen de métricas de usuarios activos
print("=== USUARIOS ACTIVOS POR PERÍODO ===\n")
print(f"DAU promedio: {dau['daily_users'].mean():.0f} usuarios/día")
print(f"WAU promedio: {wau['weekly_users'].mean():.0f} usuarios/semana")
print(f"MAU promedio: {mau['monthly_users'].mean():.0f} usuarios/mes \n")

# Estadísticas descriptivas de usuarios activos por periodo
display(dau['daily_users'].describe())    # Métricas estadísticas de usuarios activos diarios (DAU)
display(wau['weekly_users'].describe())   # Métricas estadísticas de usuarios activos semanales (WAU)
display(mau['monthly_users'].describe())  # Métricas estadísticas de usuarios activos mensuales (MAU)
```

```python
# Identificar pico y valle de MAU
mau_max = mau.loc[mau['monthly_users'].idxmax()]
mau_min = mau.loc[mau['monthly_users'].idxmin()]

print(f"Pico: {mau_max['month'].strftime('%B %Y')} - {mau_max['monthly_users']:,} MAU")
print(f"Valle: {mau_min['month'].strftime('%Y-%m')} - {mau_min['monthly_users']:,} MAU")
print(f"Diferencia: {((mau_max['monthly_users'] - mau_min['monthly_users']) / mau_min['monthly_users'] * 100):.1f}%")
```

### 2.1.3. Duración de cada sesión

```python
# Calcular la duración de cada sesión en minutos
visits['session_duration_minutes'] = (
    visits['end_ts'] - visits['start_ts']
).dt.total_seconds() / 60

# Estadísticas descriptivas de la duración de sesiones
session_duration_stats = visits['session_duration_minutes'].describe()

# Duración promedio por dispositivo
session_duration_by_device = (
    visits
    .groupby('device')['session_duration_minutes']
    .mean()
    .reset_index(name='avg_session_minutes')
)

# Duración promedio por fuente de adquisición
session_duration_by_source = (
    visits
    .groupby('source_id')['session_duration_minutes']
    .mean()
    .reset_index(name='avg_session_minutes')
    .sort_values(by='avg_session_minutes', ascending=False)
)

print("=== DURACIÓN DE SESIONES ===\n")
display(session_duration_stats) # Estadísticas descriptivas de la duración de las sesiones (en minutos)
display(session_duration_by_device) # Duración promedio de las sesiones por tipo de dispositivo
display(session_duration_by_source) # Duración promedio de las sesiones por fuente de adquisición
```

## 2.1.4. Frecuencia de retorno de usuarios

```python
# Número de sesiones por usuario
sessions_per_user = (
    visits
    .groupby('uid')
    .size()
    .reset_index(name='total_sessions')
)

# Categorizar usuarios según frecuencia
def categorize_user_frequency(sessions):
    if sessions == 1:
        return 'Usuario único (1 sesión)'
    elif sessions <= 5:
        return 'Usuario ocasional (2-5 sesiones)'
    elif sessions <= 20:
        return 'Usuario regular (6-20 sesiones)'
    else:
        return 'Usuario frecuente (20+ sesiones)'

sessions_per_user['user_category'] = sessions_per_user['total_sessions'].apply(categorize_user_frequency)

# Distribución de usuarios por frecuencia
user_frequency_distribution = (
    sessions_per_user['user_category']
    .value_counts()
    .reset_index(name='usuarios')
)
user_frequency_distribution['porcentaje'] = (
    user_frequency_distribution['usuarios'] / len(sessions_per_user) * 100
).round(2)

# Tasa de retorno
returning_users = sessions_per_user[sessions_per_user['total_sessions'] > 1].shape[0]
total_users = sessions_per_user.shape[0]
return_rate = (returning_users / total_users) * 100

# Timeline de sesiones por usuario
user_sessions_timeline = (
    visits
    .groupby('uid')
    .agg(
        first_session=('start_ts', 'min'),
        last_session=('start_ts', 'max'),
        total_sessions=('start_ts', 'count')
    )
    .reset_index()
)

# Calcular días activos y frecuencia
user_sessions_timeline['days_active'] = (
    user_sessions_timeline['last_session'] - user_sessions_timeline['first_session']
).dt.days

# Solo usuarios recurrentes (más de 1 sesión)
returning_users_timeline = user_sessions_timeline[user_sessions_timeline['total_sessions'] > 1].copy()
returning_users_timeline['avg_days_between_sessions'] = (
    returning_users_timeline['days_active'] / (returning_users_timeline['total_sessions'] - 1)
)

# Resumen de frecuencia de retorno
print("=== FRECUENCIA DE RETORNO DE USUARIOS ===\n")
print(f"Total de usuarios únicos: {total_users:,}")
print(f"Usuarios que regresaron: {returning_users:,}")
print(f"Tasa de retorno: {return_rate:.2f}%")
print(f"\nPromedio de días entre sesiones: {returning_users_timeline['avg_days_between_sessions'].mean():.1f} días")
print(f"Mediana de días entre sesiones: {returning_users_timeline['avg_days_between_sessions'].median():.1f} días\n")

display(sessions_per_user['total_sessions'].describe()) # Estadísticas descriptivas del número de sesiones por usuario
display(user_frequency_distribution) # Distribución de usuarios según la frecuencia de sesiones
```

## 2.2. ANÁLISIS DE VENTAS

Se evalúa el comportamiento de compra de los usuarios: el momento en que comienzan a comprar, el número de pedidos en distintos períodos, el tamaño promedio de compra y el valor total generado por cliente (LTV). Estas métricas permiten medir el impacto real de los usuarios en los ingresos del negocio.

### 2.2.1. Tiempo hasta la primera compra

```python
# Unir visits con orders para encontrar la primera visita de cada usuario
first_visit_per_user = (
    visits
    .groupby('uid')['start_ts']
    .min()
    .reset_index(name='first_visit')
)

first_order_per_user = (
    orders
    .groupby('uid')['buy_ts']
    .min()
    .reset_index(name='first_order')
)

# Combinar primera visita con primera compra
conversion_timeline = first_visit_per_user.merge(
    first_order_per_user, 
    on='uid', 
    how='left'
)

# Calcular tiempo hasta primera compra (en días)
conversion_timeline['days_to_first_purchase'] = (
    conversion_timeline['first_order'] - conversion_timeline['first_visit']
).dt.total_seconds() / (60 * 60 * 24)

# Solo usuarios que compraron
buyers_timeline = conversion_timeline[conversion_timeline['first_order'].notna()].copy()

# Estadísticas de tiempo hasta primera compra
print("=== TIEMPO HASTA LA PRIMERA COMPRA ===\n")
print(f"Total de usuarios que visitaron: {len(conversion_timeline):,}")
print(f"Total de usuarios que compraron: {len(buyers_timeline):,}")
print(f"Tasa de conversión: {(len(buyers_timeline)/len(conversion_timeline)*100):.2f}%")
print(f"\nPromedio de días hasta primera compra: {buyers_timeline['days_to_first_purchase'].mean():.1f} días")
print(f"Mediana de días hasta primera compra: {buyers_timeline['days_to_first_purchase'].median():.1f} días\n")

display(buyers_timeline['days_to_first_purchase'].describe())

# Distribución del tiempo hasta primera compra
conversion_categories = pd.cut(
    buyers_timeline['days_to_first_purchase'],
    bins=[0, 1, 7, 30, 90, float('inf')],
    labels=['Mismo día', '2-7 días', '1-4 semanas', '1-3 meses', '3+ meses']
)

conversion_distribution = (
    conversion_categories
    .value_counts()
    .reset_index(name='usuarios')
    .sort_values(by='index')
)
conversion_distribution['porcentaje'] = (
    conversion_distribution['usuarios'] / len(buyers_timeline) * 100
).round(2)

print("\nDISTRIBUCIÓN DEL TIEMPO HASTA PRIMERA COMPRA:")
display(conversion_distribution) # Distribución de usuarios según el número de conversiones
```

### 2.2.2. Número de pedidos por usuario y período

```python
# Crear columna de mes de compra
orders['order_month'] = orders['buy_ts'].astype('datetime64[M]')

# Total de pedidos por usuario
orders_per_user = (
    orders
    .groupby('uid')
    .size()
    .reset_index(name='total_orders')
)

# Categorizar usuarios según frecuencia de compra
def categorize_buyer_frequency(orders_count):
    if orders_count == 1:
        return 'Comprador único'
    elif orders_count <= 3:
        return 'Comprador ocasional (2-3)'
    elif orders_count <= 10:
        return 'Comprador regular (4-10)'
    else:
        return 'Comprador frecuente (10+)'

orders_per_user['buyer_category'] = orders_per_user['total_orders'].apply(categorize_buyer_frequency)

# Distribución de compradores por frecuencia
buyer_frequency_distribution = (
    orders_per_user['buyer_category']
    .value_counts()
    .reset_index(name='usuarios')
)
buyer_frequency_distribution['porcentaje'] = (
    buyer_frequency_distribution['usuarios'] / len(orders_per_user) * 100
).round(2)

# Pedidos por mes
orders_per_month = (
    orders
    .groupby('order_month')
    .size()
    .reset_index(name='total_orders')
)

# Pedidos por día
orders['order_date'] = orders['buy_ts'].dt.date
orders_per_day = (
    orders
    .groupby('order_date')
    .size()
    .reset_index(name='total_orders')
)

print("=== NÚMERO DE PEDIDOS POR USUARIO Y PERÍODO ===\n")
print(f"Total de compradores únicos: {len(orders_per_user):,}")
print(f"Total de pedidos: {len(orders):,}")
print(f"Promedio de pedidos por usuario: {orders_per_user['total_orders'].mean():.2f}")
print(f"\nPromedio de pedidos por mes: {orders_per_month['total_orders'].mean():.0f}")
print(f"Promedio de pedidos por día: {orders_per_day['total_orders'].mean():.0f}\n")

display(orders_per_user['total_orders'].describe()) # Estadísticas descriptivas del número de pedidos por usuario
display(buyer_frequency_distribution) # Distribución de compradores según la frecuencia de pedidos
display(orders_per_month.head(10)) # Primeros registros del número de pedidos agregados por mes
```

### 2.2.3. Tamaño promedio de compra

```python
# Estadísticas del valor de pedido
order_value_stats = orders['revenue'].describe()

# Valor promedio por dispositivo
# Primero unir orders con visits para obtener información de dispositivo
orders_with_device = orders.merge(
    visits[['uid', 'device', 'start_ts']],
    on='uid',
    how='left'
)

# Filtrar solo visitas que ocurrieron antes o el mismo día de la compra
orders_with_device = orders_with_device[
    orders_with_device['start_ts'] <= orders_with_device['buy_ts']
]

# Tomar el dispositivo de la última sesión antes de comprar
orders_with_device = (
    orders_with_device
    .sort_values('start_ts')
    .groupby(['uid', 'buy_ts', 'revenue'])
    .tail(1)
    .reset_index(drop=True)
)

avg_order_by_device = (
    orders_with_device
    .groupby('device')['revenue']
    .agg(['mean', 'count'])
    .reset_index()
    .rename(columns={'mean': 'avg_revenue', 'count': 'total_orders'})
)

# Valor promedio por fuente de adquisición
# Obtener fuente de cada usuario (primera visita)
user_source = (
    visits
    .sort_values('start_ts')
    .groupby('uid')['source_id']
    .first()
    .reset_index()
)

orders_with_source = orders.merge(user_source, on='uid', how='left')

avg_order_by_source = (
    orders_with_source
    .groupby('source_id')['revenue']
    .agg(['mean', 'count'])
    .reset_index()
    .rename(columns={'mean': 'avg_revenue', 'count': 'total_orders'})
    .sort_values('avg_revenue', ascending=False)
)

# Distribución de tamaño de compra
order_size_categories = pd.cut(
    orders['revenue'],
    bins=[0, 1, 5, 10, 50, float('inf')],
    labels=['Micro ($0-1)', 'Pequeña ($1-5)', 'Media ($5-10)', 'Grande ($10-50)', 'Premium ($50+)']
)

order_size_distribution = (
    order_size_categories
    .value_counts()
    .reset_index(name='pedidos')
    .sort_values(by='index')
)
order_size_distribution['porcentaje'] = (
    order_size_distribution['pedidos'] / len(orders) * 100
).round(2)

print("=== TAMAÑO PROMEDIO DE COMPRA ===\n")
print(f"Valor promedio de pedido: ${orders['revenue'].mean():.2f}")
print(f"Valor mediano de pedido: ${orders['revenue'].median():.2f}")
print(f"Valor total de ingresos: ${orders['revenue'].sum():,.2f}\n")

display(order_value_stats) # Estadísticas descriptivas del valor de los pedidos
display(avg_order_by_device) # Valor promedio del pedido por tipo de dispositivo
display(avg_order_by_source) # Valor promedio del pedido por fuente de adquisición
display(order_size_distribution) # Distribución de pedidos según el tamaño de compra
```

### 2.2.4. LTV (Lifetime Value) - Valor de vida del cliente

```python
# LTV por usuario: ingresos totales que genera cada cliente
ltv_per_user = (
    orders
    .groupby('uid')['revenue']
    .sum()
    .reset_index(name='ltv')
)

# Estadísticas de LTV
ltv_stats = ltv_per_user['ltv'].describe()

# LTV por fuente de adquisición
ltv_by_source = ltv_per_user.merge(user_source, on='uid', how='left')

ltv_summary_by_source = (
    ltv_by_source
    .groupby('source_id')['ltv']
    .agg(['mean', 'median', 'sum', 'count'])
    .reset_index()
    .rename(columns={
        'mean': 'ltv_promedio',
        'median': 'ltv_mediana',
        'sum': 'ltv_total',
        'count': 'usuarios'
    })
    .sort_values('ltv_promedio', ascending=False)
)

# LTV por dispositivo
ltv_by_device = ltv_per_user.merge(
    visits.groupby('uid')['device'].first().reset_index(),
    on='uid',
    how='left'
)

ltv_summary_by_device = (
    ltv_by_device
    .groupby('device')['ltv']
    .agg(['mean', 'median', 'sum', 'count'])
    .reset_index()
    .rename(columns={
        'mean': 'ltv_promedio',
        'median': 'ltv_mediana',
        'sum': 'ltv_total',
        'count': 'usuarios'
    })
)

# LTV acumulado por cohorte mensual
# Identificar mes de primera compra de cada usuario
first_purchase_month = (
    orders
    .groupby('uid')['buy_ts']
    .min()
    .reset_index()
)
first_purchase_month['first_purchase_month'] = first_purchase_month['buy_ts'].astype('datetime64[M]')

# Unir con todas las compras
orders_with_cohort = orders.merge(
    first_purchase_month[['uid', 'first_purchase_month']],
    on='uid',
    how='left'
)

# Calcular edad de cohorte (meses desde primera compra)
orders_with_cohort['cohort_age'] = (
    (orders_with_cohort['order_month'] - orders_with_cohort['first_purchase_month']) 
    / pd.Timedelta(days=30)
).round().astype(int)

# LTV acumulado por cohorte
ltv_by_cohort = (
    orders_with_cohort
    .groupby(['first_purchase_month', 'cohort_age'])
    .agg(
        revenue=('revenue', 'sum'),
        usuarios=('uid', 'nunique')
    )
    .reset_index()
)

ltv_by_cohort['ltv_per_user'] = ltv_by_cohort['revenue'] / ltv_by_cohort['usuarios']

# Tabla pivote de LTV por cohorte
ltv_cohort_pivot = (
    ltv_by_cohort
    .pivot_table(
        index='first_purchase_month',
        columns='cohort_age',
        values='ltv_per_user',
        aggfunc='sum'
    )
    .cumsum(axis=1)
    .round(2)
)

print("=== LTV (LIFETIME VALUE) ===\n")
print(f"LTV promedio por usuario: ${ltv_per_user['ltv'].mean():.2f}")
print(f"LTV mediano por usuario: ${ltv_per_user['ltv'].median():.2f}")
print(f"LTV total: ${ltv_per_user['ltv'].sum():,.2f}\n")

display(ltv_stats) # Estadísticas descriptivas del LTV de los usuarios
display(ltv_summary_by_source) # LTV promedio por fuente de adquisición
display(ltv_summary_by_device) # LTV promedio por tipo de dispositivo
display(ltv_cohort_pivot.iloc[:, :7]) # LTV acumulado por cohorte durante los primeros 6 meses
```

## 2.3. ANÁLISIS DE MARKETING

Se analizan los gastos de marketing por fuente y a lo largo del tiempo, el costo de adquisición de clientes (CAC) y la rentabilidad de las inversiones (ROMI). Este análisis permite comparar la eficiencia de las distintas fuentes de tráfico y su contribución a los ingresos.

### 2.3.1. Gasto en marketing

```python
# Gasto total
total_marketing_spend = costs['costs'].sum()

# Gasto por fuente de adquisición
spend_by_source = (
    costs
    .groupby('source_id')['costs']
    .sum()
    .reset_index(name='total_spend')
    .sort_values('total_spend', ascending=False)
)

# Porcentaje del gasto por fuente
spend_by_source['percentage'] = (
    spend_by_source['total_spend'] / total_marketing_spend * 100
).round(2)

# Gasto a lo largo del tiempo (mensual)
costs['month'] = costs['dt'].astype('datetime64[M]')

spend_by_month = (
    costs
    .groupby('month')['costs']
    .sum()
    .reset_index(name='monthly_spend')
)

# Gasto por fuente a lo largo del tiempo
spend_by_source_month = (
    costs
    .groupby(['month', 'source_id'])['costs']
    .sum()
    .reset_index(name='spend')
)

# Estadísticas de gasto diario
daily_spend_stats = costs['costs'].describe()

print("=== GASTO EN MARKETING ===\n")
print(f"Gasto total en marketing: ${total_marketing_spend:,.2f}")
print(f"Gasto promedio mensual: ${spend_by_month['monthly_spend'].mean():,.2f}")
print(f"Gasto promedio diario: ${costs.groupby('dt')['costs'].sum().mean():,.2f}\n")

display(spend_by_source) # Gasto por fuente de adquisición
display(spend_by_month.head(5)) # Gasto mensual (primeros 5 meses)
display(daily_spend_stats) # Estadísticas de gasto diario
```

### 2.3.2. CAC (Customer Acquisition Cost) por fuente

```python
# Número de usuarios adquiridos por fuente
# Un usuario se atribuye a la fuente de su primera visita
users_by_source = (
    visits
    .sort_values('start_ts')
    .groupby('uid')['source_id']
    .first()
    .reset_index()
)

users_acquired_by_source = (
    users_by_source
    .groupby('source_id')
    .size()
    .reset_index(name='users_acquired')
)

# Calcular CAC: gasto total por fuente / usuarios adquiridos por fuente
cac_by_source = spend_by_source.merge(
    users_acquired_by_source,
    on='source_id',
    how='left'
)

cac_by_source['cac'] = (
    cac_by_source['total_spend'] / cac_by_source['users_acquired']
).round(2)

cac_by_source = cac_by_source.sort_values('cac', ascending=False)

# CAC solo para usuarios que compraron (compradores adquiridos)
buyers_by_source = user_source.copy()

buyers_acquired_by_source = (
    buyers_by_source
    .groupby('source_id')
    .size()
    .reset_index(name='buyers_acquired')
)

cac_buyers_by_source = spend_by_source.merge(
    buyers_acquired_by_source,
    on='source_id',
    how='left'
)

cac_buyers_by_source['cac_buyer'] = (
    cac_buyers_by_source['total_spend'] / cac_buyers_by_source['buyers_acquired']
).round(2)

cac_buyers_by_source = cac_buyers_by_source.sort_values('cac_buyer', ascending=False)

print("=== CAC (CUSTOMER ACQUISITION COST) ===\n")
print(f"CAC promedio (todos los usuarios): ${cac_by_source['cac'].mean():.2f}")
print(f"CAC promedio (solo compradores): ${cac_buyers_by_source['cac_buyer'].mean():.2f}\n")

display(cac_by_source) # CAC por fuente (todos los usuarios)
display(cac_buyers_by_source[['source_id', 'buyers_acquired', 'cac_buyer']]) # CAC por fuente (solo compradores)
```

### 2.3.3. ROMI (Return on Marketing Investment)

```python
# Ingresos por fuente de adquisición
revenue_by_source = (
    orders_with_source
    .groupby('source_id')['revenue']
    .sum()
    .reset_index(name='total_revenue')
)

# Combinar ingresos, gastos y CAC
romi_by_source = spend_by_source.merge(
    revenue_by_source,
    on='source_id',
    how='left'
)

romi_by_source = romi_by_source.merge(
    cac_buyers_by_source[['source_id', 'buyers_acquired', 'cac_buyer']],
    on='source_id',
    how='left'
)

romi_by_source = romi_by_source.merge(
    ltv_summary_by_source[['source_id', 'ltv_promedio']],
    on='source_id',
    how='left'
)

# Calcular ROMI: (ingresos - gastos) / gastos
romi_by_source['romi'] = (
    (romi_by_source['total_revenue'] - romi_by_source['total_spend']) / 
    romi_by_source['total_spend']
).round(2)

# Calcular ROI porcentual
romi_by_source['roi_percentage'] = (romi_by_source['romi'] * 100).round(2)

# Relación LTV/CAC
romi_by_source['ltv_cac_ratio'] = (
    romi_by_source['ltv_promedio'] / romi_by_source['cac_buyer']
).round(2)

# Ordenar por ROMI
romi_by_source = romi_by_source.sort_values('romi', ascending=False)

# ROMI a lo largo del tiempo (mensual)
# Primero, obtener ingresos por mes y fuente
revenue_by_month_source = (
    orders_with_source
    .groupby(['order_month', 'source_id'])['revenue']
    .sum()
    .reset_index(name='revenue')
)

# Combinar con gastos mensuales por fuente
romi_monthly = spend_by_source_month.merge(
    revenue_by_month_source,
    left_on=['month', 'source_id'],
    right_on=['order_month', 'source_id'],
    how='left'
).fillna(0)

romi_monthly['romi'] = (
    (romi_monthly['revenue'] - romi_monthly['spend']) / romi_monthly['spend']
).round(2)

# ROMI acumulado por fuente
romi_cumulative_by_source = (
    romi_monthly
    .sort_values(['source_id', 'month'])
    .groupby('source_id')
    .apply(lambda x: x.assign(
        cumulative_spend=x['spend'].cumsum(),
        cumulative_revenue=x['revenue'].cumsum()
    ))
    .reset_index(drop=True)
)

romi_cumulative_by_source['cumulative_romi'] = (
    (romi_cumulative_by_source['cumulative_revenue'] - romi_cumulative_by_source['cumulative_spend']) / 
    romi_cumulative_by_source['cumulative_spend']
).round(2)

print("=== ROMI (RETURN ON MARKETING INVESTMENT) ===\n")
print(f"ROMI promedio: {romi_by_source['romi'].mean():.2f} ({romi_by_source['roi_percentage'].mean():.0f}%)")
print(f"Relación LTV/CAC promedio: {romi_by_source['ltv_cac_ratio'].mean():.2f}")
print(f"Ingresos totales: ${revenue_by_source['total_revenue'].sum():,.2f}")
print(f"Gastos totales: ${total_marketing_spend:,.2f}")
print(f"Beneficio neto marketing: ${(revenue_by_source['total_revenue'].sum() - total_marketing_spend):,.2f}\n")

display(romi_by_source) # ROMI completo por fuente
display(romi_monthly.head(15)) # ROMI mensual (primeras 15 filas)
display(romi_cumulative_by_source[romi_cumulative_by_source['source_id'] == 1].head(10)) # ROMI acumulado Fuente 1 (ejemplo)
```

```python
# Primero, calcular CAC por dispositivo
# Obtener el dispositivo de cada usuario (primera visita)
user_device = (
    visits
    .sort_values('start_ts')
    .groupby('uid')['device']
    .first()
    .reset_index()
)

# Calcular gasto estimado por dispositivo (proporcional a usuarios)
users_by_device = user_device['device'].value_counts().reset_index()
users_by_device.columns = ['device', 'total_users']

# Calcular proporción de gasto por dispositivo
total_spend = costs['costs'].sum()
users_by_device['user_proportion'] = users_by_device['total_users'] / users_by_device['total_users'].sum()
users_by_device['estimated_spend'] = users_by_device['user_proportion'] * total_spend

# Obtener compradores por dispositivo
buyers_by_device = user_device.merge(
    orders[['uid']].drop_duplicates(),
    on='uid',
    how='inner'
)
buyers_count_by_device = buyers_by_device['device'].value_counts().reset_index()
buyers_count_by_device.columns = ['device', 'buyers']

# Calcular CAC por dispositivo
cac_by_device = users_by_device.merge(buyers_count_by_device, on='device')
cac_by_device['cac'] = cac_by_device['estimated_spend'] / cac_by_device['buyers']

# Calcular LTV por dispositivo
ltv_by_device_data = user_device.merge(
    orders.groupby('uid')['revenue'].sum().reset_index(),
    on='uid',
    how='inner'
)
ltv_summary_by_device = (
    ltv_by_device_data
    .groupby('device')['revenue']
    .agg(['sum', 'count'])
    .reset_index()
)
ltv_summary_by_device.columns = ['device', 'revenue', 'buyers']
ltv_summary_by_device['ltv'] = ltv_summary_by_device['revenue'] / ltv_summary_by_device['buyers']

# Renombrar para consistencia con el gráfico
ltv_by_device = ltv_summary_by_device.copy()

# Ahora el gráfico funcionará
fig, ax = plt.subplots(figsize=(12, 6))

x = np.arange(len(cac_by_device))
width = 0.35

# Crear barras agrupadas
bars1 = ax.bar(x - width/2, cac_by_device['cac'], width, label='CAC', 
               color='#d62728', edgecolor='black', linewidth=1.2)
bars2 = ax.bar(x + width/2, ltv_by_device['ltv'], width, label='LTV', 
               color='#2ca02c', edgecolor='black', linewidth=1.2)

# Título y etiquetas
ax.set_title('Comparación CAC vs LTV por Dispositivo', fontsize=16, fontweight='bold', pad=20)
ax.set_xlabel('Dispositivo', fontsize=12, fontweight='bold')
ax.set_ylabel('USD', fontsize=12, fontweight='bold')

# Formato de ejes
ax.set_xticks(x)
ax.set_xticklabels(cac_by_device['device'], fontsize=11)
ax.tick_params(axis='y', labelsize=11)

# Leyenda
ax.legend(fontsize=11, loc='upper right')

# Grilla sutil
ax.grid(axis='y', alpha=0.3, linestyle='--', linewidth=0.7)

# Agregar valores en las barras CAC
for i, (bar, row_cac) in enumerate(zip(bars1, cac_by_device.itertuples())):
    height = bar.get_height()
    x_center = bar.get_x() + bar.get_width()/2.
    
    # Valor CAC arriba de la barra
    ax.text(x_center, height + 0.1,
            f'${row_cac.cac:.2f}',
            ha='center', va='bottom', fontsize=11, fontweight='bold', color='black')
    
    # Compradores dentro de la barra (CAC)
    ax.text(x_center, height * 0.70,
            'Compradores:',
            ha='center', va='bottom', fontsize=12, fontweight='bold', color='white')
    ax.text(x_center, height * 0.62,
            f'{row_cac.buyers:,}',
            ha='center', va='top', fontsize=12, fontweight='bold', color='white')
    
    # Gasto dentro de la barra (CAC)
    ax.text(x_center, height * 0.40,
            'Gasto:',
            ha='center', va='bottom', fontsize=12, fontweight='bold', color='white')
    ax.text(x_center, height * 0.32,
            f'${row_cac.estimated_spend:,.0f}',
            ha='center', va='top', fontsize=12, fontweight='bold', color='white')

# Agregar valores en las barras LTV
for i, (bar, row_ltv) in enumerate(zip(bars2, ltv_by_device.itertuples())):
    height = bar.get_height()
    x_center = bar.get_x() + bar.get_width()/2.
    
    # Valor LTV arriba de la barra
    ax.text(x_center, height + 0.1,
            f'${row_ltv.ltv:.2f}',
            ha='center', va='bottom', fontsize=12, fontweight='bold', color='black')
    
    # Compradores dentro de la barra (LTV)
    ax.text(x_center, height * 0.70,
            'Compradores:',
            ha='center', va='bottom', fontsize=12, fontweight='bold', color='white')
    ax.text(x_center, height * 0.62,
            f'{row_ltv.buyers:,}',
            ha='center', va='top', fontsize=12, fontweight='bold', color='white')
    
    # Revenue dentro de la barra (LTV)
    ax.text(x_center, height * 0.40,
            'Revenue:',
            ha='center', va='bottom', fontsize=12, fontweight='bold', color='white')
    ax.text(x_center, height * 0.32,
            f'${row_ltv.revenue:,.0f}',
            ha='center', va='top', fontsize=12, fontweight='bold', color='white')

plt.tight_layout()
plt.show()
```

## 3.2. Métricas Utilizadas para las Recomendaciones

### ¿En qué métricas nos enfocamos y por qué?

**1. Ratio LTV/CAC - Métrica principal de decisión**
- **Qué mide**: Cuántos dólares de valor genera cada dólar invertido en adquisición
- **Por qué es la más importante**: Combina eficiencia (CAC) y retorno (LTV) en un solo indicador de sostenibilidad
- **Benchmark saludable**: ≥ 3.0 (cada USD invertido debe generar USD 3+ en retorno)
- **Nuestro resultado global**: 4.66 (aparentemente saludable, pero con extrema variación por fuente)

**2. ROMI (Return on Marketing Investment) - Validación de rentabilidad**
- **Qué mide**: (Ingresos - Gastos) / Gastos → ganancia o pérdida directa
- **Por qué complementa**: Valida si el ratio LTV/CAC se traduce en rentabilidad real en el período analizado
- **Nuestro resultado global**: -7% (pérdida neta de USD 23,017)

**3. CAC (Customer Acquisition Cost) - Eficiencia**
- **Qué mide**: Gasto total / compradores adquiridos
- **Nuestro resultado global**: USD 1.48 promedio (muy bajo vs benchmark e-commerce USD 20-50)

**4. LTV (Lifetime Value) - Valor del cliente**
- **Qué mide**: Ingresos totales por cliente
- **Nuestro resultado global**: USD 6.90 promedio
- **Limitante crítica**: 82.77% son compradores únicos → LTV depende casi exclusivamente de primera compra
